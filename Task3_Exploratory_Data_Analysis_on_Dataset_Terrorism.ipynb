Step 4: Data Visualization

Country with most attacks

```python
plt.subplots(figsize=(20, 6))
sns.barplot(df['country_txt'].value_counts().index[:40],df['country_txt'].value_counts()[:40].values)
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.title('Country with the most attacks', fontsize =20)
plt.show()
```

```python
print("Country with the most attacks:",df['country_txt'].value_counts().idxmax())
```

Region by most attack

```python
plt.subplots(figsize = (20, 6))
sns.countplot('region_txt', data = df,palette='icefire',edgecolor=sns.color_palette('dark',7))
plt.xticks(rotation=20)
plt.xlabel('region')
plt.title('Terrorist Attacks Regionwise', fontsize = 18)
plt.show()
```

Least affected country

```python
plt.subplots(figsize=(20, 6))
sns.barplot(df['country_txt'].value_counts().index[-40:],df['country_txt'].value_counts()[-40:].values,palette='mako')
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.title('Country with the least attacks', fontsize = 20)
plt.show()
```

```python
print("Least targetted Country is:", df['country_txt'].value_counts().index[-1])
```

Attack Year VS Region

```python
dff_region=pd.crosstab(df.iyear,df.region)
dff_region.plot(color=sns.color_palette('Set2',12))
fig=plt.gcf()
plt.title("Terrorist Attack Year VS Region", fontsize=20)
fig.set_size_inches(20,7)
plt.show()
```

Attack Year VS Target Type

```python
dff_region=pd.crosstab(df.iyear, df.targtype1)
dff_region.plot(color=sns.color_palette('Set2',12))
fig=plt.gcf()
plt.title("Terrorist Attack Year VS Target Type", fontsize=20)
fig.set_size_inches(15,8)
plt.show()
```
